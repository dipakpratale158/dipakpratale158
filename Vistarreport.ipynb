```python
import requests
from bs4 import BeautifulSoup

# URL of the page to scrape
url = 'https://in21.leadsquared.com/ActivityManagement?ec=247'

# Make a GET request to fetch the page content
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Parse the HTML content
    soup = BeautifulSoup(response.text, 'html.parser')

    # Find all anchor tags (<a>)
    links = soup.find_all('a')

    # Filter and print links that start with the desired URL
    for link in links:
        href = link.get('href')
        if href and href.startswith('https://developerapp-pdf.leadsquared.com'):
            print(href)
else:
    print(f"Failed to retrieve the page. Status code: {response.status_code}")
```
