```python
# //////////////////////////////////////////////dont useseven copy past final filter monthly data process dealerwise
import xlwings as xw
import os

def copy_sheet_data(source_file, destination_file, source_sheet_name, destination_sheet_name):
    try:
        # Check if source file exists
        if not os.path.isfile(source_file):
            raise FileNotFoundError(f"Source file does not exist: {source_file}")

        # Check if destination file exists
        if not os.path.isfile(destination_file):
            raise FileNotFoundError(f"Destination file does not exist: {destination_file}")

        # Open the source workbook and select the source sheet
        source_wb = xw.Book(source_file)
        if source_sheet_name not in [sheet.name for sheet in source_wb.sheets]:
            raise ValueError(f"Source sheet '{source_sheet_name}' does not exist in the source workbook.")
        source_ws = source_wb.sheets[source_sheet_name]

        # Read the source data into a DataFrame
        source_data = source_ws.range('A1').expand().value

        # Open the destination workbook and select or create the destination sheet
        destination_wb = xw.Book(destination_file)
        if destination_sheet_name in [sheet.name for sheet in destination_wb.sheets]:
            destination_ws = destination_wb.sheets[destination_sheet_name]
        else:
            destination_ws = destination_wb.sheets.add(name=destination_sheet_name)

        # Clear existing data in the destination sheet
        destination_ws.range('A1').expand().clear_contents()

        # Write the data to the destination sheet
        destination_ws.range('A1').value = source_data

        # Save and close the workbooks
        destination_wb.save()
        source_wb.close()
        destination_wb.close()

        print("Data copied successfully!")

    except Exception as e:
        print(f"An error occurred: {e}")

# Define file paths and sheet names
source_file = r'C:\Users\thotnr\Desktop\vbadealerwise.xlsm'
destination_file = r'C:\Users\thotnr\Desktop\summury report\Exchange MC data summary report - AP TG WB.xlsx'
source_sheet_name = 'Processed Data -Vehicle-Status'  # Name of the sheet to copy from
destination_sheet_name = 'Processed Data -Vehicle-Status'  # Name of the sheet to paste into

copy_sheet_data(source_file, destination_file, source_sheet_name, destination_sheet_name)



```

```python


# /////////////////////////////////////////////// 6 six: please dealerwise anylysis vba echo and absolute formating then edite formula then run this code six copy pastr
# fil monthly report     after compliting chek bu filter for mhp 
import xlwings as xw
import os

def copy_sheets_to_multiple_destinations(source_file, destination_files, sheets_mappings):
    try:
        # Check if source file exists
        if not os.path.isfile(source_file):
            raise FileNotFoundError(f"Source file does not exist: {source_file}")

        # Open the source workbook
        source_wb = xw.Book(source_file)

        for destination_file, sheets_mapping in sheets_mappings.items():
            # Check if destination file exists
            if not os.path.isfile(destination_file):
                raise FileNotFoundError(f"Destination file does not exist: {destination_file}")

            # Open the destination workbook
            destination_wb = xw.Book(destination_file)

            for source_sheet_name, destination_sheet_name in sheets_mapping.items():
                if source_sheet_name not in [sheet.name for sheet in source_wb.sheets]:
                    print(f"Source sheet '{source_sheet_name}' does not exist in the source workbook.")
                    continue

                # Select the source sheet
                source_ws = source_wb.sheets[source_sheet_name]
                source_data = source_ws.range('A1').expand().value

                if destination_sheet_name in [sheet.name for sheet in destination_wb.sheets]:
                    destination_ws = destination_wb.sheets[destination_sheet_name]
                else:
                    destination_ws = destination_wb.sheets.add(name=destination_sheet_name)

                # Clear existing data in the destination sheet
                destination_ws.range('A1').expand().clear_contents()

                # Write the data to the destination sheet
                destination_ws.range('A1').value = source_data

                # Autofit columns
                destination_ws.range('A1').expand().columns.autofit()

                # Apply borders
                destination_ws.range('A1').expand().api.Borders.Weight = 2  # xlThin = 2

                # Apply header formatting
                header_range = destination_ws.range('A1').expand('right')
                header_range.api.Font.Bold = True
                header_range.color = (169, 208, 142)  # Light Green background color

            # Save and close the destination workbook
            destination_wb.save()
            destination_wb.close()

        # Close the source workbook
        source_wb.close()

        print("Data copied and formatted successfully!")

    except Exception as e:
        print(f"An error occurred: {e}")

# Define file paths
source_file = r'C:\Users\thotnr\Desktop\vbadealerwise.xlsm'
destination_files = [
    r'C:\Users\thotnr\Desktop\summury report\Exchange MC data summary report - AP TG WB.xlsx',
    r'C:\Users\thotnr\Desktop\summury report\Exchange MC data summary report - MAH,TN,VID.xlsx'
]

# Define the mapping for each destination file
sheets_mappings = {
    r'C:\Users\thotnr\Desktop\summury report\Exchange MC data summary report - AP TG WB.xlsx': {
        'Dealerwise Analysis': 'AP Dealerwise Analysis',
        'APvehicalstatus': 'AProcessed Data -Vehicle-Status',
        'APdealerwise': 'AP Processed Data -Dealer-wise',
        'APbrokerperformance': 'AP Broker Performance',
        'DSE PerformanceMHP':'AP DSE Performance'
    },
    r'C:\Users\thotnr\Desktop\summury report\Exchange MC data summary report - MAH,TN,VID.xlsx': {
        'Dealerwise Analysis mhp': 'MHP Dealerwise Analysis',
        'MHPvehicalstatus': 'MHrocessed Data -Vehicle-Status',
        'MHPdealerwise': 'MHP Processed Data -Dealer-wise',
        'MHPbrokerperformance': 'MHP Broker Performance',
        'DSE Performance':'MHP DSE Performance'
    }
}

copy_sheets_to_multiple_destinations(source_file, destination_files, sheets_mappings)
```

```python
# ////////////////////// fifth monthly report in dsesheet add download Q column =IF(O3<>"","Yes","No")  , vba slicer click helper m n 
# // vba vehical bid add j colum make uniq key =I27&M2

import xlwings as xw
import pandas as pd

# Define the file path
file_path = r"C:\Users\thotnr\Desktop\vbadealerwise.xlsm"

# Open the workbook
wb = xw.Book(file_path)

def filter_and_paste_dse(sheet_name, column_name, regions, target_sheet):
    """Filter data from the DSE sheet, paste into A to N columns of the target sheet, and add formulas in P to T columns."""
    ws = wb.sheets[sheet_name]
    
    # Load data from the DSE sheet
    data = ws.range('A1').expand().options(pd.DataFrame, header=1).value
    print("Data columns:", data.columns)  # Print the column names for debugging

    # Filter the data based on regions
    filtered_data = data[data[column_name].isin(regions)]
    
    # Access the target sheet
    target_ws = wb.sheets[target_sheet]
    
    # Clear existing data by resetting the entire range to empty values
    target_ws.range('A1:Z10000').value = ''  # Assumes maximum of 100000 rows and 26 columns to clear
    
    # Paste filtered data starting from column A (A:N)
    target_ws.range('B1').value = filtered_data.iloc[:, :13]  # Paste columns A to N (0-indexed, so A:N is 0:13)

    # Write 'Unique Key' in P1 and the formula in P2
    target_ws.range('P1').value = 'Unique Key'
    target_ws.range('P2:P' + str(filtered_data.shape[0] + 1)).formula = '=H2&M2'  # Formula for concatenating columns H and M

    # Write 'Total Enquiries in App' in Q1 and the formula in Q2
    target_ws.range('Q1').value = 'Total Enquiries in App'
    target_ws.range('Q2:Q' + str(filtered_data.shape[0] + 1)).formula = '=COUNTIF(vehicalstatusprocessfile!D:D, P2)'  # Formula for counting in vehical_status

    # Write 'Enquiries with minimum 1 bid' in R1 and the formula in R2
    target_ws.range('R1').value = 'Enquiries with minimum 1 bid'
    target_ws.range('R2:R' + str(filtered_data.shape[0] + 1)).formula = '=COUNTIF(helper!N:N, P2)'  # Formula for counting bids in helper sheet

    # Write 'Total Bids' in S1 and the formula in S2
    target_ws.range('S1').value = 'Total Bids'
    target_ws.range('S2:S' + str(filtered_data.shape[0] + 1)).formula = "=COUNTIF('vehicles bids'!J:J,P2)"  # Formula for counting in vehicles bids

    # Write 'Bids per Enquiry' in T1 and the formula in T2
    target_ws.range('T1').value = 'Bids per Enquiry'
    target_ws.range('T2:T' + str(filtered_data.shape[0] + 1)).formula = '=IFERROR(S2/R2, 0)'  # Formula for bids per enquiry

    # Write 'Platform' in U1 and the formula in U2
    target_ws.range('U1').value = 'Platform'
    target_ws.range('U2:U' + str(filtered_data.shape[0] + 1)).formula = '=IF(XLOOKUP(N2, dse!M:M, dse!O:O) = 0, "", XLOOKUP(N2, dse!M:M, dse!O:O))'  # Adjust based on your sheet structure

    # Write 'Version' in V1 and the formula in V2
    target_ws.range('V1').value = 'Version'
    target_ws.range('V2:V' + str(filtered_data.shape[0] + 1)).formula = '=IF(XLOOKUP(N2, dse!M:M, dse!p:p) = 0, "", XLOOKUP(N2, dse!M:M, dse!p:p))'  # Adjust based on your sheet structure

    # Write 'Download' in W1 and the formula in W2
    target_ws.range('W1').value = 'Download'
    target_ws.range('W2:W' + str(filtered_data.shape[0] + 1)).formula = '=XLOOKUP(N2, dse!M:M, dse!Q:Q)'  # Adjust based on your sheet structure

    # Write 'ID' in A1 and the sequence formula for A2:A (IDs)
    target_ws.range('A1').value = 'ID'
    target_ws.range('A2:A' + str(filtered_data.shape[0] + 1)).formula = '=ROW()-ROW($A$1)'  # Alternative for SEQUENCE to generate IDs


# Apply the filter and pasting for the regions and target sheets
filter_and_paste_dse('dse', 'Region Name', ['MAH', 'TN', 'VID'], 'DSE Performance')
filter_and_paste_dse('dse', 'Region Name', ['AP', 'TG', 'WB'], 'DSE PerformanceMHP')

# Save and close the workbook
wb.save()
# wb.close()  # Uncomment if you want to close the workbook

print("DSE data has been filtered, pasted, and formulas added successfully.")
```

```python
import pandas as pd

# Define the path to the Excel file
file_path = r"C:\Users\thotnr\Desktop\vbadealerwise.xlsm"

# Load the data from the Excel file
df = pd.read_excel(file_path, sheet_name='Dealerwise Analysis')  # Adjust sheet_name accordingly

# Create the pivot table for "Total enquiries in App" and "Vehicles acquired"
pivot = df.pivot_table(
    index=["Region", "Hub"],
    values=["Total enquiries in App", "Vehicles acquired"],
    aggfunc="sum",
    margins=True
)

# Calculate the percentage of "Vehicles Acquired" relative to "Total enquiries in App"
pivot['Percentage Vehicles Acquired'] = (pivot['Vehicles acquired'] / pivot['Total enquiries in App']) * 100

# Format the percentage column to show as a percentage with 2 decimal places
pivot['Percentage Vehicles Acquired'] = pivot['Percentage Vehicles Acquired'].map(lambda x: f"{x:.2f}%")

# Define the output path for the Excel file
output_path = r"C:\Users\thotnr\Desktop\Combined_Report.xlsx"

# Write both the original data and the pivot table to the same Excel file with different sheets
with pd.ExcelWriter(output_path) as writer:
    df.to_excel(writer, sheet_name='Original Data', index=False)
    pivot.to_excel(writer, sheet_name='Pivot Table', index=True)

print(f"Data and pivot table exported to {output_path}")
```
